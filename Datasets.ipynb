```python
import sklearn
from sklearn.datasets import load_digits
digits = load_digits()

print(type(digits))
print(digits.feature_names)

x = digits.data
y = digits.target
print(x)
print(y)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x,y)
print(dt.score(x,y))
```

```python
from sklearn.cluster import KMeans
kt = KMeans()
kt.fit (x,y)
print(kt.score(x,y))
```
